```python
from num2words.num2words_CH_main import num2words
import pandas as pd
from detect_number_spans import detect_number_spans
```

```python
def convert_numbers(text: str,dialect) -> str:
    spans = detect_number_spans(text)

    # Work from end to start so indices stay valid when we modify the string
    spans.sort(key=lambda s: s.start, reverse=True)
    out = text

    for span in spans:
        number = span.text

        if span.kind == "NUMBER":
            # normal cardinal number → words
            print(number)
            number = num2words(number, lang=dialect)
        elif span.kind == "YEAR":
            if int(number) > 1999:
                number = num2words(number, lang=dialect)
            elif len(number) == 4:
                number = num2words(number[:2], lang=dialect) + " " + num2words(number[2:], lang=dialect)
        elif span.kind == "ZIP":
            if len(number) == 4:
                number = num2words(number[:2], lang=dialect) + " " + num2words(number[2:], lang=dialect)

        elif span.kind == "PHONE":
            # TODO
            pass
        elif span.kind == "CAR_PLATE":
            # TODO
            pass
        elif span.kind == "ORDINAL":
            number = num2words(number[:-1], lang=dialect, ordinal=True)
        elif span.kind == "MONEY":
            # custom money formatting, e.g. "10.–" → "zehn Franken"
            # Here you'd parse amount + currency.
            pass
        elif span.kind == "MODEL":
            # model numbers normally stay as-is
            pass

        out = out[:span.start] + number + out[span.end:]

    return out
```

```python
convert_numbers("Basel 12.3 4410 Liestal ist die Postleitzahl von Liestal.", dialect="ch_bs")
```
